# Part 1: Temporal Patterns

```python
focuscrimes = set(['WEAPON LAWS', 'PROSTITUTION', 'DRIVING UNDER THE INFLUENCE', 'ROBBERY', 'BURGLARY', 'ASSAULT', 'DRUNKENNESS', 'DRUG/NARCOTIC', 'TRESPASS', 'LARCENY/THEFT', 'VANDALISM', 'VEHICLE THEFT', 'STOLEN PROPERTY', 'DISORDERLY CONDUCT'])
```

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

#Reading the data and setting up the essentials
df = pd.read_csv('Police_Department_Incident_Reports__Historical_2003_to_May_2018.csv')
df['dt'] = df.Date.astype(str) +" "+ df.Time.astype(str)
df['Datetime']=df['dt'].apply(lambda x: dt.datetime.strptime(x, "%m/%d/%Y %H:%M"))
del df['dt']
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Week'] = df['Datetime'].dt.week
df['Date'] = df['Datetime'].dt.date
df['Hour'] = df['Datetime'].dt.hour
df['Weekday'] = df['Datetime'].dt.weekday
df = df.loc[ df["Year"] <= 2017,] #We are assuming, that we are still asked to investigate data between 2003 and 2017.
df = df.loc[ df["Year"] >= 2013,]
df = df[df["Category"].isin(focuscrimes)]
```

* *Weekly patterns*. Basically, we'll forget about the yearly variation and just count up what happens during each weekday. [Here's what my version looks like](https://raw.githubusercontent.com/suneman/socialdataanalysis2020/master/files/weekdays.png). Some things make sense - for example `drunkenness` and the weekend. But there are some aspects that were surprising to me. Check out `prostitution` and mid-week behavior, for example!?

```python
#Plot the results for weekdays
x_pos = 0
y_pos = 0

fig, axs = plt.subplots(7, 2,figsize=(15,15))

#Count the incidents in each focus crime and group by day of the week.
df_temp = df[['Category','DayOfWeek','IncidntNum']].groupby(['Category','DayOfWeek']).size().unstack()
df_temp = df_temp[['Monday', 'Tuesday', 'Wednesday','Thursday','Friday','Saturday','Sunday']]

for i in range(len(df_temp)):
    
    if(x_pos > 6):
        x_pos = 0
        y_pos = 1
        
    
    axs[x_pos, y_pos].bar(df_temp.iloc[i,].index,df_temp.iloc[i,])
    axs[x_pos, y_pos].set_title(df_temp.iloc[i,].name)
    axs[x_pos, y_pos].set(xlabel='Weekday', ylabel='Count')
    x_pos = x_pos+1
fig.tight_layout(pad=3.0)
```

As stated in the sub question, it can be seen, that arrests for 'DRUNKNESS' are more likely to happen during the weekend. As a prostitute, you are more likely to be arrested in the mid-week. 'LARCENY/THEFT' have a little peak during the weekend. Both 'DRUNKNESS' and 'LARCENY/THEFT' make sense, but we cannot explain, why a prostitute is more likely to be arrested in the mid-week.

* *The months*. We can also check if some months are worse by counting up number of crimes in Jan, Feb, ..., Dec. Did you see any surprises there?

```python
#Plot the results for the crimes for months
x_pos = 0
y_pos = 0

fig, axs = plt.subplots(7, 2,figsize=(15,15))

df_temp = df[['Category','Month','IncidntNum']].groupby(['Category','Month']).size().unstack()

for i in range(len(df_temp)):
    
    if(x_pos > 6):
        x_pos = 0
        y_pos = 1
        
    
    axs[x_pos, y_pos].bar(df_temp.iloc[i,].index,df_temp.iloc[i,])
    axs[x_pos, y_pos].set_title(df_temp.iloc[i,].name)
    axs[x_pos, y_pos].set(xlabel='Month', ylabel='Count')
    x_pos = x_pos+1
fig.tight_layout(pad=3.0)
```

In the start of the year and in the cold months (January and February), there is a tendency to arrest more people for 'PROSTITUTION'. One could see, that there may be a seasonel trend in 'PROSTITUTION'. In a way 'DRUG/NARCOTIC' seems to have negative tendency over the year.

* *The 24 hour cycle*. We'll can also forget about weekday and simply count up the number of each crime-type that occurs in the entire dataset from midnight to 1am, 1am - 2am ... and so on. Again: Give me a couple of comments on what you see. 

```python
#Plot the results for the crimes for months
x_pos = 0
y_pos = 0

fig, axs = plt.subplots(7, 2,figsize=(15,15))

df_temp = df[['Category','Hour','IncidntNum']].groupby(['Category','Hour']).size().unstack()

for i in range(len(df_temp)):
    
    if(x_pos > 6):
        x_pos = 0
        y_pos = 1
        
    
    axs[x_pos, y_pos].bar(df_temp.iloc[i,].index,df_temp.iloc[i,])
    axs[x_pos, y_pos].set_title(df_temp.iloc[i,].name)
    axs[x_pos, y_pos].set(xlabel='Hour', ylabel='Count')
    x_pos = x_pos+1
fig.tight_layout(pad=3.0)
```

Most crimes happen between 15 and 20 hours from midnight, except from 'DRUNKNESS', 'PROSTITUTION' and 'DRIVING UNDER THE INFLUENCE', who happen often at night. There is a wierd peak for 'PROSTITUTION' at 12 - maybe lunch break? The most arrest for 'DISORDERLY CONDUCT' happen between 6-9, when most people tend to go to work.

* *Hours of the week*. But by looking at just 24 hours, we may be missing some important trends that can be modulated by week-day, so let's also check out the 168 hours of the week. So let's see the number of each crime-type Monday night from midninght to 1am, Monday night from 1am-2am - all the way to Sunday night from 11pm to midnight.


```python
df['HourOfWeek'] = df['Hour']+(df['Weekday']*24)
#Plot the results for the crimes for months
x_pos = 0
y_pos = 0

fig, axs = plt.subplots(7, 2,figsize=(15,15))

df_temp = df[['Category','HourOfWeek','IncidntNum']].groupby(['Category','HourOfWeek']).size().unstack()

for i in range(len(df_temp)):
    
    if(x_pos > 6):
        x_pos = 0
        y_pos = 1
        
    
    axs[x_pos, y_pos].bar(df_temp.iloc[i,].index,df_temp.iloc[i,])
    axs[x_pos, y_pos].set_title(df_temp.iloc[i,].name)
    axs[x_pos, y_pos].set(xlabel='HourOfWeek', ylabel='Count')
    x_pos = x_pos+1
fig.tight_layout(pad=3.0)
```

Having the Hour of the week allow detection of different kind of patterns. The most 'PROSTITUTION' arrests happen Thursday in the early hours. There is a peak for 'TRESSPASS' on Wednesday, when people tend to go to work.

* Simpson's paradox is hard to explain. Come up with your own example - or find one on line.

On Saturday, you solved 7 out of 8 attempted problems, but your friend solved 2 out of 2. You had solved more problems, but your friend pointed out that he was more accurate, 7/8 < 2/2. 
	
On Sunday, you only attempted 2 problems and got 1 correct. Your friend got 5 out of 8 problems correct. Your friend gloated once again, since 1/2 < 5/8.

However, the competition is about the one who solved more accurately over the weekend, not on individual days. Overall, you have solved 8 out of 10 problems whereas your friend has solved 7 out of 10 problems. Thus, despite your friend solving a higher proportion of problems on each day, you actually won the challenge by solving the higher proportion for the entire weekend!

# Part 3: Generating important plot types

* First, let's make a jitter-plot (that is, code up something like **Figure 2-1** from DAOST from scratch), but based on SF Police data. My hunch from inspecting the file is that the police-folks might be a little bit lazy in noting down the **exact** time down to the second. So choose a crime-type and a suitable time interval (somewhere between a month and 6 months depending on the crime-type) and create a jitter plot of the arrest times during a single hour (like 13-14, for example). So let time run on the $x$-axis and create vertical jitter.

```python
df['Day'] = df['Datetime'].dt.day
df_temp = df.loc[df["Category"]=='PROSTITUTION']
df_temp = df_temp.loc[(df_temp["Year"]==2016) & (df_temp["Month"]<7)]
df_temp = df_temp.loc[df_temp["Hour"]==12]

plt.scatter(df_temp["Day"], np.random.normal(1, 1, len(df_temp)),marker = "o", facecolors='none', edgecolors='r')
plt.title('Jitter plot over Prostitution in 2016, first six months and time = 12')
plt.xlabel('Dates')
plt.ylabel('Randomness')
plt.grid()
plt.show()
```

In the jitterplot above there occur many observations at the same days of the months. This can be visualized by a jitterplot, which point out the many observations.

* Now for some histograms (please create a crime-data based versions of the plot-type shown in DAOST **Figure 2-2**). (I think the GPS data could be fun to understand from this perspective.) 
  * This time, pick two crime-types with different geographical patterns **and** a suitable time-interval for each (you want between 1000 and 10000 points in your histogram)
  * Then take the latitude part of the GPS coordinates for each crime and bin the latitudes so that you have around 50 bins across the city of SF. You can use your favorite method for binning. I like `numpy.histogram`. This function gives you the counts and then you do your own plotting. 

```python
df_temp = df.loc[(df["Year"]>=2013) & (df["Year"]<=2017)]
df_temp_Pros = df_temp.loc[df_temp["Category"]=='PROSTITUTION']
df_temp_Weapon = df_temp.loc[df_temp["Category"]=='WEAPON LAWS']
plt.hist(df_temp_Pros["Y"],bins='auto')
plt.xlabel('Latitude')
plt.ylabel('Counts')
plt.title('Latitude for prostitution')
plt.grid()
plt.show()

plt.hist(df_temp_Weapon["Y"],bins=50)
plt.xlabel('Latitude')
plt.ylabel('Counts')
plt.title('Latitude for weapon law')
plt.grid()
plt.show()
```

* Next up is using the plot-type shown in **Figure 2-4** from DAOST, but with the data you used to create Figure 2.1. To create the kernel density plot, you can either use `gaussian_kde` from `scipy.stats` ([for an example, check out this stackoverflow post](https://stackoverflow.com/questions/4150171/how-to-create-a-density-plot-in-matplotlib)) or you can use [`seaborn.kdeplot`](https://seaborn.pydata.org/generated/seaborn.kdeplot.html).

```python
from scipy.stats import gaussian_kde
df_temp = df.loc[df['Category'] == 'PROSTITUTION']
df_temp = df_temp.loc[df_temp['Month'] < 7]
df_temp = df_temp.loc[df_temp['Hour'] == 12]
df_temp = df_temp.loc[df_temp['Year'] == 2016]
df_temp['Day'] = df_temp['Datetime'].dt.day


data = df_temp['Day']
density = gaussian_kde(data)
xs = np.linspace(0,31,200)
density.covariance_factor = lambda : .25
density._compute_covariance()
plt.plot(xs,density(xs))
plt.title('Kernel Density Plot')
plt.ylabel('Density')
plt.xlabel('Dates')
plt.grid()
plt.show()
```

This density plot should look familar as the histo gram, but with more smoothness.

* Finally, grab 25 random timepoints from the dataset (of 1000-10000 original data) you've just plotted and create a version of Figure 2-4 based on the 25 data points. Does this shed light on why I think KDEs can bee misleading? 

```python
df_temp['Random'] = np.random.normal(1,1,len(df_temp))
df_temp = df_temp.sort_values(by=['Random'])
df_temp = df_temp[:25]
data = df_temp['Day']
density = gaussian_kde(data)
xs = np.linspace(0,31,200)
density.covariance_factor = lambda : .25
density._compute_covariance()
plt.plot(xs,density(xs))
plt.title('Kernel Density Plot with 25 random samples')
plt.ylabel('Density')
plt.xlabel('Dates')
plt.grid()
plt.show()
```

Here we have a KDE with only 25 samples, and the KDE looks very smooth and nice, but only with 25 samples. It looks like almost perfect data, which could be misleading. With a histogram it would not have been the case, because it would have shown the counts.
